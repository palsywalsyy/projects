```python
# Load Pustaka/Library yang dibutuhkan

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Tampilkan Boston Housing dataset (.csv file)

df = pd.read_csv(r'C:\Users\shaff\Downloads\Boston_housing.csv')
df
```

```python
# Buat DataFrame yang lebih kecil dengan kolom yang tidak terdapat CHAS, NOX, B, dan LSTAT

df.drop(columns=['CHAS', 'NOX', 'B', 'LSTAT'], inplace=True)
```

```python
# Periksa 7 baris terakhir dari DataFrame yang baru

df.tail(7)
```

```python
#Gambarkan visualisasi histogram dari semua variable/kolom pada DataFrame yang baru

plt.hist(['CRIM','ZN','INDUS','RM','AGE','DIS','RAD','TAX','PTRATIO', 'PRICE'])
plt.show()
```

```python
# Gambarkan semua histogram menggunakan for loop. Tambahkan nama yang unik pada plot

histograms = ['CRIM', 'ZN', 'INDUS', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'PRICE']
for col in histograms:
    plt.figure()
    plt.hist(df[col]) 
    plt.title("Histogram " + col) 
    plt.ylabel("Frekuensi "  + col)
    plt.xlabel(col)   
    plt.show() 
    plt.tight_layout()
```

```python
# Buat scatter plot dari crime rate versus price

plt.figure(figsize=(15, 8))
plt.scatter(df['CRIM'], df['PRICE'])
plt.title("Plot of Crime Rate vs. Price",\
         fontsize=20)
plt.xlabel("Crime Rate", fontsize=16)
plt.ylabel("Price", fontsize=16)
plt.show()
```

```python
# Buat Plot log10(crime) versus price.

plt.figure(figsize=(15,8))
plt.title('Plot of Crime Rate vs Price', fontsize=20)
plt.xlabel('Log Crime Rate', fontsize=16)
plt.ylabel('Price', fontsize=16)
plt.semilogx(df['CRIM'], df['PRICE'])
plt.show()
```
